Below you'll find code for the UDMIS model of interest. Strictly speaking, for the purposes of performing the Monte Carlo simulation we can simply compute the required energy differences using two calls to the energy function. However, we can often save significant computation time by writing a specialised function for the energy difference that avoids unnecessary calculations. Simply put, calculating the energy of a configuration of occupations requires you to iterate over all vertices in the graph, but the energy change associated with a single-occupation-flip can be computed just from values of the connected vertices.

Define initial graph

```python
samples = [line.split(' ')[0] for line in open('task2_data.dat').readlines()]
sol = max(set(samples), key = samples.count)
word_counts = Counter(samples)
color_map = []
G = nx.from_numpy_matrix(udmis.find_edges())
string = [int(d) for d in sol[::-1]]
for node in range(udmis.num_vertices):
    if string[node]==0:
        color_map.append('black')
    else:
        color_map.append('salmon')
nx.draw_circular(G,node_color=color_map)
```

Define simulation parameters

```python
N = 10000
t = np.arange(N+1)
T_i = 100
T_f = 0.01
```

Define simulation functions

```python
def sim(T):
    strings=[]
    energies=[]
    for t in range(N+1):
        # take a look at the abstract_udmis.py file to see how mc_step works
        temp = T[t]
        E = udmis.mc_step(T=temp)
        if t % 100 == 0:
            string = udmis.occupations
            strings.append(string.tolist())
            energies.append(E)
    return strings,energies
```

Define plotting functions

```python
def animate(strings):
    fig, ax = plt.subplots()
    G = nx.from_numpy_matrix(udmis.find_edges())

    def update(frame):
        ax.clear()
        color_map = []
        st = strings[frame]
        for node in range(len(st)):
            if not st[node]:
                color_map.append('black')
            else:
                color_map.append('salmon')
        nx.draw_circular(G,node_color=color_map)
        step = frame*100
        ax.annotate('Step = %d'%step, xy =(1, 1),
                 xytext =(1,1))

    ani = animation.FuncAnimation(fig,update,frames=int(N/100),repeat=False)
    ani.save('graph.gif',dpi=80, writer=PillowWriter(fps=5))

def plot_energies(energies,label):
    plt.plot(energies,label=label)
```

Run simulation with linear interpolation


```python
#linear
lin = (1-(t/N))*T_i + (t/N)*T_f
strings_lin,energies_lin = sim(lin)
```

Run simulation with exponential

```python
#original
exp1 = T_i * ((T_f/T_i) ** (t/N))
exp2 = T_i * ((T_f/T_i) ** (2*(t/N)))
exp3 = T_i * ((T_f/T_i) ** (3*(t/N)))
strings_exp1,energies_exp1 = sim(exp1)
strings_exp2,energies_exp2 = sim(exp2)
strings_exp3,energies_exp3 = sim(exp3)
animate(strings_exp1)
```

Cosine interpolation

```python
co = (1.0 - np.cos(t/N*np.pi))*0.5
c = (1-co)*T_i + co*T_f
strings_c,energies_c = sim(c)
```

Plot and compare

```python
plt.figure()
plt.plot(lin,label='linear')
plt.plot(c,label='cos')
plt.plot(exp1,label='exp1')
plt.plot(exp2,label='exp2')
plt.plot(exp3,label='exp3')
plt.legend()
plt.show()
```

Plot and compare minimization path

```python
plt.figure()
plot_energies(energies_lin,'linear')
plot_energies(energies_c,'cos')
plot_energies(energies_exp1,'exp1')
plot_energies(energies_exp2,'exp2')
plot_energies(energies_exp3,'exp3')
plt.legend()
plt.show()
```
