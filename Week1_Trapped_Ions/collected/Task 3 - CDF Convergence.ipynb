# CDF Convergence

```julia
include("task_functions.jl")
```

```julia
# Define our variables
N = 8
low_depth = 1
high_depth = 512
flipon = true

# Wavefunctions
low_depth_psi = run(N, low_depth, flipon)
high_depth_psi = run(N, high_depth, flipon)

# Probabilities
low_depth_prob = get_Probs(low_depth_psi, N)
high_depth_prob = get_Probs(high_depth_psi, N)

# Convert Probabilities to CDF
h = 0.0001
low_depth_cdf = take_derivative(low_depth_prob, h)
high_depth_cdf = take_derivative(high_depth_prob, h)
```

```julia
histogram(low_depth_cdf, nbins=100)
```

```julia
histogram(high_depth_cdf)
```
