```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_o = range(2025,2061,1)
y_orig = [0.403, 0.7, 0.996, 1.292, 1.589, 1.885, 2.176, 2.466, 2.757, 3.047, 3.338, 3.664, 3.991, 4.318, 4.644, 4.971, 5.271, 5.57, 5.87, 6.17, 6.469, 6.765, 7.062, 7.358, 7.654, 7.951, 8.259, 8.568, 8.876, 9.185, 9.494, 9.776, 10.057, 10.338, 10.62, 10.901]

x_r = range(2025,2055)
y_2orig = [1.541, 2.087, 2.632, 3.178, 3.721, 4.262, 4.804, 5.346, 5.888, 6.429, 6.971, 7.482, 7.99, 8.499, 9.007, 9.515, 10.035, 10.556, 11.077, 11.598, 12.119, 12.637, 13.154, 13.672, 14.189, 14.706, 15.243, 15.781, 16.32, 16.858]

y_o = [10**x for x in y_orig]
y_or = [10**x for x in y_2orig]
```

```python
fqec = lambda pl: (4*np.log(10**0.5*(0.0168)/pl)/np.log(0.01/0.0168) + 1)**(-2)
```

```python
pls = [float('1.0e-{}'.format(x)) for x in range(2,18)]
nps = [int(1.0/fqec(pl)) for pl in pls]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot(pls,nps)
ax.set_yscale('log')
```

```python
# Transition Metals 
s = 1
p = 3
d = 5
f = 7
g = 9
h = 11
i = 13

ccpv2z = 6*s + 5*p + 3*d + 1*f
ccpv5z = 9*s + 8*p + 6*d + 4*f + 3*g + 2*h + i

print('ccpv2z:',ccpv2z, ccpv2z*2)
print('ccpv5z:',ccpv5z, ccpv5z*2)
```

```python
y_new1 = [ y*fqec(1.0e-2)/fqec(1.0e-18) for y in y_o]
y_new2 = [ y*fqec(1.0e-2)/fqec(1.0e-18) for y in y_or]
```
